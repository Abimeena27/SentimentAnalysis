```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from bs4 import BeautifulSoup
import re
import requests
```

```python
import pandas as pd
import numpy as np
```

```python
tokenizer=AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
tokens = tokenizer.encode('I feel i am the GOD', return_tensors='pt')
```

```python
tokens
```

```python
tokenizer.decode(tokens[0])
```

```python
res=model(tokens)
```

```python
res
```

```python
sc=torch.argmax(res.logits)+1
```

```python
int(sc)
```

```python
content=requests.get('https://www.yelp.com/biz/nothing-bundt-cakes-daly-city')
```

```python
content.text
```

```python
txt=BeautifulSoup(content.text,'html.parser')
```

```python
txt
```

```python
regex=re.compile('.*comment*.')
```

```python
regex
```

```python
results=txt.find_all('p',{'class':regex})
```

```python
results
```

```python
writings=[result.text for result in results]
```

```python
writings
```

```python
comment=np.array(writings)
```

```python
data=pd.DataFrame(comment, columns=['Comments'])
```

```python
data.head()
```

```python
data.shape
```

```python
def sentiment_analyzer(text):
    tokens = tokenizer.encode(text, return_tensors='pt')
    res=model(tokens)
    score=torch.argmax(res.logits)+1
    return int(score)
```

```python
score=sentiment_analyzer(data['Comments'].iloc[0])
score
```

```python
scores=[]
for c in data['Comments']:
    score=sentiment_analyzer(c)
    scores.append(score)
data['Sentiment-Score']=scores
```
